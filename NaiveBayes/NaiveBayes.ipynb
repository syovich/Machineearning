```python
from sklearn.datasets import load_digits, load_breast_cancer

dataset = load_breast_cancer()
X = dataset.data
y = dataset.target

print X
```

```python
from sklearn.cross_validation import cross_val_score
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```

```python
gnb = GaussianNB()
bnb = BernoulliNB()
mnb = MultinomialNB()

fraction = 0.8
L = len(X)
count = int(fraction * L)
X_train = X[:count]
y_train = y[:count]

X_test = X[count:]
y_test = y[count:]
```

```python
y_pred = gnb.fit(X_train, y_train).predict(X_test)
print sum(y_test == y_pred) / float(len(y_test))
```

```python
import numpy as np
alphas = np.logspace(-4, -.5, 30)

scores = list()
scores_std = list()

for alpha in alphas:
    gnb.alpha = alpha
    this_scores = cross_val_score(gnb, X, y, n_jobs=1)
    scores.append(np.mean(this_scores))
    scores_std.append(np.std(this_scores))
```

```python
print this_scores
print scores
print scores_std
```

```python
score = cross_val_score(gnb, X, y)
print score
score = cross_val_score(bnb, X, y)
print score
score = cross_val_score(mnb, X, y)
print score
```

```python
digit = load_digits()
Xd = digit.data
yd = digit.target
score = cross_val_score(gnb, Xd, yd)
print score
score = cross_val_score(bnb, Xd, yd)
print score
score = cross_val_score(mnb, Xd, yd)
print score
```
